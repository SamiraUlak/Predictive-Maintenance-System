```python
import pandas as pd
df=pd.read_csv("iotSensor_data.csv")
print(df.shape)
df.head()
```

```python
df.isnull().sum()
print(df.dtypes)          
```

```python
df.duplicated().sum()
```

```python
failure_cols = ['Tool wear Failure', 'Heat dissipition failure',
                'power failure', 'overstrain failure', 'random failure']

for col in failure_cols:
    failure = (df[col] == 1).sum()
    non_failure = (df[col] == 0).sum()
    print(f"{col}: Failure={failure}, Non-Failure={non_failure}")

# Check your new Failure column separately
print(f"\nFinal Failure column: Failure={( df['Failure'] == 1).sum()}, Non-Failure={(df['Failure'] == 0).sum()}")
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=['UDI', 'Product ID'])
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
df['Type']=le.fit_transform(df["Type"])
```

```python
print(df.dtypes)
print(df.shape)
print(df.head())
```

```python
from sklearn.utils import resample

# Separate failure and non-failure
df_failure = df[df['Failure'] == 1]
df_non_failure = df[df['Failure'] == 0]

print(f"Before upsampling:")
print(f"Failure: {len(df_failure)}")
print(f"Non-Failure: {len(df_non_failure)}")

# Upsample failure
df_failure_upsampled = resample(df_failure,
                                replace=True,
                                n_samples=len(df_non_failure),
                                random_state=42)

# Combine
df_upsampled = pd.concat([df_non_failure, df_failure_upsampled])

print(f"\nAfter upsampling:")
print(df_upsampled['Failure'].value_counts())
```

```python
df_upsampled.head()
```

```python
# For ML model
X = df_upsampled[['Type', 'Air temperature [K]', 'Process temperature [K]',
                  'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']]
y = df_upsampled['Failure']

print(X.shape)
print(y.value_counts())
```

```python
df_upsampled=df_upsampled.sample(frac=1, random_state=42)
df_upsampled=df_upsampled.reset_index(drop=True)
```
